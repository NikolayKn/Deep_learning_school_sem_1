# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

## Применение линейных моделей (3 балла)

1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.

2) С помощью кроссвалидации или разделения на train/valid выборку протестируйте разные значения гиперпараметра C и выберите лучший (можно тестировать С=100, 10, 1, 0.1, 0.01, 0.001) по метрике ROC-AUC.

Если вы разделяете на train/valid, то используйте LogisticRegressionCV. Он сам при вызове .fit() подберет параметр С. (не забудьте передать scroing='roc_auc', чтобы при кроссвалидации сравнивались значения этой метрики, и refit=True, чтобы при потом модель обучилась на всем датасете с лучшим параметром C).


(более сложный вариант) Если вы будете использовать кроссвалидацию, то преобразования данных и LogisticRegression нужно соединить в один Pipeline с помощью make_pipeline, как это делалось во втором семинаре. Потом pipeline надо передать в GridSearchCV. Для one-hot-encoding'a можно испльзовать комбинацию LabelEncoder + OneHotEncoder (сначала превращаем строчки в числа, а потом числа првращаем в one-hot вектора.)

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import warnings
warnings.filterwarnings("ignore")
```

```python
# YOUR CODE

X = data[feature_cols]
y = data[target_col]

cat_pipe = make_pipeline(OneHotEncoder())
num_pipe = make_pipeline(StandardScaler())

feature_prep = ColumnTransformer([
    ('category preprocessing', cat_pipe, cat_cols),
    ('numeric preprocessing', num_pipe, num_cols)])

pipe = Pipeline([
    ('preproccessing', feature_prep),
    ('model', LogisticRegression())
])

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)


parameters = {
    'model__C': [100, 10, 1, 0.1, 0.01, 0.001],
    'model__penalty':['l1', 'l2', 'elasticnet']
    
    }

gs = GridSearchCV(estimator=pipe, param_grid=parameters, n_jobs=-1, cv=5, refit=True, scoring='roc_auc', verbose=10)
gs.fit(X_train,y_train)

print('best params: ', gs.best_params_)
```

```python
gs.best_score_
```

```python
roc_auc_score(y_valid, gs.best_estimator_.predict_proba(X_valid)[:, 1])
```

## Применение градиентного бустинга (2 балла)

```python
# YOUR CODE
from catboost import CatBoostClassifier, Pool

X = data[feature_cols]
y = data[target_col]

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3,stratify=y, random_state=42)

model = CatBoostClassifier(verbose=True)

model.fit(X_train,y_train, cat_features=cat_cols)
```

```python
roc_auc_score(y_valid, model.predict_proba(X_valid)[:,1],average='macro')
```

### Grid search

```python
best = {}
```

```python
#my grid search

from  itertools import product
from sklearn.model_selection import KFold
from tqdm import tqdm

kf = KFold(n_splits=5)

train_pool = Pool(X, label = y, cat_features=cat_cols)

def get_parameters(grid):
    names = list(grid.keys())
    comb = list(product(*list(grid.values())))
    return [dict(zip(names, x)) for x in comb]

parameter_grid = {
    'iterations':[1000],
    'learning_rate':np.linspace(0.01,0.5,2),
    'depth':[5, 6],
    'l2_leaf_reg': [2, 3],
    'model_size_reg': [0.45, 0.5],
    'loss_function':['Logloss'],
    'verbose':[0]

}

grid = get_parameters(parameter_grid)

for param in tqdm(grid):
    param_score = np.array([])
    for train_index, test_index in kf.split(X):
#         print(param)
        tr_pool = train_pool.slice(train_index)
        val_pool = train_pool.slice(test_index)
        
        model = CatBoostClassifier(**param)
        model.fit(tr_pool, eval_set = [(val_pool)], early_stopping_rounds = 50)
        score = roc_auc_score(y.iloc[test_index], model.predict_proba(X.iloc[test_index])[:,1], average='micro')
        param_score = np.append(param_score, score)
        
    best[param_score.mean()] = param
        
        
        
```

```python
def get_best(res):
    score = sorted(res.items(), key=lambda x: -x[0])[0][0]
    param = sorted(res.items(), key=lambda x: -x[0])[0][1]
    return score, param

best_score, best_param = get_best(best)

print('best_score', best_score, '\nbest_param', best_param)
```

```python
model = CatBoostClassifier(silent=True, eval_metric='AUC', cat_features=cat_cols)

model.grid_search({
    'iterations': np.arange(200,250,300),
    'l2_leaf_reg': [1,2,3,4],
    'random_seed': [0],
    'depth': [4,5,6,7],
    'learning_rate': np.linspace(0.06,0.1, 10),
    }, 
    X_train, 
    y_train, plot=True, refit=True)
```

```python
best_params = model.get_all_params()
ad_param = ['iterations','l2_leaf_reg', 'random_seed', 'depth', 'learning_rate']
my_param = {key: model.get_all_params()[key] for key in ad_param}
print(my_param)
```

```python
roc_auc_score(y_valid, model.predict_proba(X_valid)[:,1],average='micro')
```
